```
from snowflake.snowpark.context import get_active_session
from snowflake.snowpark.functions import col

session = get_active_session()

df_raw = session.table("IRIS_DB.PUBLIC.IRIS")
```

```
df_clean = df_raw.select(
    col("SEPAL_LENGTH").cast("double").alias("SEPAL_LENGTH"),
    col("SEPAL_WIDTH").cast("double").alias("SEPAL_WIDTH"),
    col("PETAL_LENGTH").cast("double").alias("PETAL_LENGTH"),
    col("PETAL_WIDTH").cast("double").alias("PETAL_WIDTH"),
    col("VARIETY")
)
```

```
df_clean.columns
```

```
from snowflake.ml.modeling.ensemble import RandomForestClassifier

model = RandomForestClassifier(label_cols="VARIETY")
model.fit(df_clean)
```

```
model.predict(df_clean).show()
```

```
model.predict(
    session.create_dataframe(
        [[5.1, 3.5, 1.4, 0.2]],
        schema=["SEPAL_LENGTH","SEPAL_WIDTH","PETAL_LENGTH","PETAL_WIDTH"]
    )
).show()
```
